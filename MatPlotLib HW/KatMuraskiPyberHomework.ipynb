```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

merge_data = pd.merge(city_data, ride_data, on="city")

# Display the data table for preview
merge_data.head()
```

```python
#city_data_cleaned=merge_table.drop_duplicates('city', keep="first")
#city_data_cleaned.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#x=total number of rides (per city)
#y=avg fares

#urban
urban_df = merge_data[merge_data["type"] == "Urban"].groupby([merge_data["city"]])
urban_x=urban_df["ride_id"].count()
urban_y=urban_df["fare"].mean()
urban_s=urban_df["driver_count"].mean()

#suburban
suburban_df = merge_data[merge_data["type"] == "Suburban"].groupby([merge_data["city"]])
suburban_x=suburban_df["ride_id"].count()
suburban_y=suburban_df["fare"].mean()
suburban_s=suburban_df["driver_count"].mean()

#rural
rural_df = merge_data[merge_data["type"] == "Rural"].groupby([merge_data["city"]])
rural_x=rural_df["ride_id"].count()
rural_y=rural_df["fare"].mean()
rural_s=rural_df["driver_count"].mean()

# Build the scatter plots for each city types
urban_scatter=plt.scatter(urban_x, urban_y, s=urban_s*10, label="Urban", marker="o", facecolors="orange", edgecolors="black", alpha=0.75)
suburban_scatter=plt.scatter(suburban_x, suburban_y, s=suburban_s*10, label="Suburban", marker="o", facecolors="skyblue", edgecolors="black", alpha=0.75)
rural_scatter=plt.scatter(rural_x, rural_y, marker="o", s=rural_s*10, label="Rural", facecolors="yellow", edgecolors="black", alpha=0.75)


# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fares($)")
plt.title("Pyber Ride Sharing Data(2016)")
plt.grid()

# Create a legend
scatter_legend=plt.legend(fontsize=8, title="City Types", loc="upper right")

# Incorporate a text label regarding circle size
plt.text(40,35, "Note: Circle size correlates with driver count per city.", fontsize=10)

# Save Figure
plt.savefig("../Images/PyberScatterPlots.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_group=merge_data.groupby(['type'])
type_fares=type_group['fare'].sum()
# Build Pie Chart

explode = (0,0,.1)
labels=["Rural","Suburban","Urban"]
colors=["yellow","lightblue","orange"]
plt.pie(type_fares, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.show()

# Save Figure
plt.savefig("../Images/PyberPieChart.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
type_group=merge_data.groupby(['type'])
total_rides=type_group['ride_id'].count()

# Build Pie Chart
explode = (0,0,.1)
labels=["Rural","Suburban","Urban"]
colors=["yellow","lightblue","orange"]
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
plt.show()

# Save Figure
plt.savefig("../Images/PyberPieTotalRides.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
type_drivers=city_data.groupby(['type'])
total_drivers=type_drivers['driver_count'].sum()


# Build Pie Charts
explode = (0,0,.1)
labels=["Rural","Suburban","Urban"]
colors=["yellow","lightblue","orange"]
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
plt.show()
# Save Figure
plt.savefig("../Images/PyberPieTotalDrivers.png")
```
